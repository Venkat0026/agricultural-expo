```python
import pandas as pd

# Step 1: Load data and preprocess date
df = pd.read_csv('nigeria_agricultural_exports.csv')

# Convert Date to datetime and extract year
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Step 2: Aggregate export value by product and year
yearly_exports = df.groupby(['Product Name', 'Year'])['Export Value'].sum().reset_index()

# Step 3: Pivot to have years as columns
pivot_exports = yearly_exports.pivot(index='Product Name', columns='Year', values='Export Value')

# Calculate percentage change year over year
pct_change = pivot_exports.pct_change(axis=1) * 100  # multiply by 100 for percentage

# Step 4: Find highest percentage increase for each product
max_increase_per_product = pct_change.max(axis=1)

# Identify the product with the highest increase
product_highest_increase = max_increase_per_product.idxmax()
highest_increase_value = max_increase_per_product.max()

print(f"The product with the highest percentage increase in export value over the years is '{product_highest_increase}' with an increase of {highest_increase_value:.2f}%.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('nigeria_agricultural_exports.csv')

# Convert 'Date' to datetime and extract Year, Month, Quarter
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter

# Define perishable products
perishable_products = ['Palm Oil', 'Rubber', 'Cassava']

# Filter dataset for perishable products
df_perishables = df[df['Product Name'].isin(perishable_products)]

# Monthly pattern: sum export values by Year, Month, Product Name
monthly_pattern = df_perishables.groupby(['Year', 'Month', 'Product Name'])['Export Value'].sum().reset_index()

# Quarterly pattern: sum export values by Year, Quarter, Product Name
quarterly_pattern = df_perishables.groupby(['Year', 'Quarter', 'Product Name'])['Export Value'].sum().reset_index()

# Display some sample results
print("Monthly Pattern Sample:")
print(monthly_pattern.head())

print("\nQuarterly Pattern Sample:")
print(quarterly_pattern.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load data and preprocess date
df = pd.read_csv('nigeria_agricultural_exports.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Step 2: Aggregate export revenue per year and product
yearly_product_revenue = df.groupby(['Year', 'Product Name'])['Export Value'].sum().reset_index()

# Aggregate total export revenue per year
total_yearly_revenue = df.groupby('Year')['Export Value'].sum().reset_index()
total_yearly_revenue = total_yearly_revenue.rename(columns={'Export Value': 'Total Export Revenue'})

# Step 3: Identify top 3 products by revenue for each year
top3_per_year = (
    yearly_product_revenue
    .sort_values(['Year', 'Export Value'], ascending=[True, False])
    .groupby('Year')
    .head(3)
)

# Step 4: Calculate share of top 3 products per year
top3_sum = top3_per_year.groupby('Year')['Export Value'].sum().reset_index()
top3_sum = top3_sum.rename(columns={'Export Value': 'Top 3 Products Revenue'})

# Merge with total yearly revenue
concentration_df = pd.merge(total_yearly_revenue, top3_sum, on='Year')

# Calculate percentage share
concentration_df['Top 3 Products Share (%)'] = (concentration_df['Top 3 Products Revenue'] / concentration_df['Total Export Revenue']) * 100

print(concentration_df)

# Step 5: Visualization of concentration trend
plt.figure(figsize=(10,6))
plt.plot(concentration_df['Year'], concentration_df['Top 3 Products Share (%)'], marker='o', linestyle='-')
plt.title('Share of Top 3 Agricultural Products in Total Export Revenue by Year')
plt.xlabel('Year')
plt.ylabel('Top 3 Products Share (%)')
plt.xticks(concentration_df['Year'].unique())
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Step 1: Load and preprocess the dataset
df = pd.read_csv('nigeria_agricultural_exports.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Step 2: Aggregate total export value per Export Country and year
yearly_country_exports = df.groupby(['Export Country', 'Year'])['Export Value'].sum().reset_index()

# Step 3: Calculate variance of yearly export values for each country
country_variance = yearly_country_exports.groupby('Export Country')['Export Value'].var().reset_index()
country_variance = country_variance.rename(columns={'Export Value': 'Yearly Export Value Variance'})

# Alternatively, to use standard deviation:
# country_variance = yearly_country_exports.groupby('Export Country')['Export Value'].std().reset_index()
# country_variance = country_variance.rename(columns={'Export Value': 'Yearly Export Value Std Dev'})

# Step 4: Sort countries by variance to find most consistent exporters (lowest variance)
most_consistent_countries = country_variance.sort_values('Yearly Export Value Variance').reset_index(drop=True)

print("Countries with most consistent export performance (lowest year-to-year variance):")
print(most_consistent_countries)
```
